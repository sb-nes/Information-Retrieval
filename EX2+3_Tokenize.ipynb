```python
import pandas as pd
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import csv
from functools import reduce
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
corp = pd.read_csv('corpus.csv', usecols = ['Name'])
```

```python
corp
```

```python
corp_data = corp['Name'].tolist()
```

```python
c = []
for data in corp_data:
    c.append(word_tokenize(data))
```

```python
token = []
for data in c:
    for i in data:
        token.append(lemmatizer.lemmatize(i))
```

```python
token_filtered = []
for data in token:
    if data.lower() in stop_words:
        continue
    if data.lower() not in token_filtered:
        if data.lower().isalnum():
            token_filtered.append(data.lower())
```

```python
token_filtered
```

```python
token_filtered.sort()
```

# TERM DOCUMENT MATRIX

```python
dictionary = {}
for data in token_filtered:
    t = []
    for i in range(51):
        l = corp.iloc[i, 0]
        if data in l.lower():
            t.append(1)
        else:
            t.append(0)   
    dictionary[f"{data}"] = t
```

```python
dictionary
```

```python
df_Dict = pd.DataFrame.from_dict(dictionary, orient ='index') 
#Converts a Python dictionary into 
```

```python
df_Dict
```

# INVERTED INDEX MATRIX

```python
inv_dictionary = {}
for data in token_filtered:
    t = []
    for i in range(51):
        l = corp.iloc[i, 0]
        if data in l.lower():
            t.append(f"d{i}")
    inv_dictionary[f"{data}"] = t
```

```python
inv_dictionary
```

```python
k = []
for id in inv_dictionary:
    if inv_dictionary[id] == []:
        k.append(id)

for delete in k:
    del inv_dictionary[delete]
#del test_dict['Mani']
```

```python
df_InvDict = pd.DataFrame.from_dict(inv_dictionary, orient ='index') 
```

```python
df_InvDict
```

# LIST OF FREQUENCY OF TOKENS

```python
token_freq = {}
for data in token_filtered:
    s=0
    for i in range(51):
        l = c[i]
        for j in l:
            if data in j.lower():
                s += 1
    token_freq[f"{data}"] = s
```

```python
token_freq
```

```python
df_tokenfreq = pd.DataFrame.from_dict(token_freq, orient ='index') 
```

```python
df_tokenfreq
```

# Save DATA to CSV

```python
with open('Token Frequency.csv', 'w', encoding="utf-8") as f:
    writer = csv.DictWriter(f, token_freq.keys())
    writer.writeheader()
    writer.writerows(token_freq)
```

```python
def editDistance(str1, str2, m, n):
    if m==0: 
        return n
    if n==0: 
        return m
    if str1[m-1] == str2[n-1]:
        return editDistance(str1, str2, m-1, n-1)
    return 1 + min(editDistance(str1, str2, m, n-1), editDistance(str1, str2, m-1, n), editDistance(str1, str2, m-1, n-1))
    # Insert, Remove, Replace
```

```python
def Correct(query, tolerance):    
    auto_correct = []
    for token in token_filtered:
        x = editDistance(query, token, len(str1), len(token))
        if x <= tolerance:
            auto_correct.append(token)
    return auto_correct
```
