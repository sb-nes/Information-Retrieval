```python
import requests
from bs4 import BeautifulSoup as bs
import csv
```

```python
r = requests.get('https://en.wikipedia.org/wiki/Nintendo')
```

```python
print(r.content)
```

```python
print(r.status_code)
```

```python
soup = bs(r.content, 'html.parser')
print(soup.prettify())
```

```python
print(soup.title)
```

```python
content = soup.find_all('p')
print(content)
```

```python
corpus_data = []
count = 0

for i in content:
    if i.text == '\n':
        continue
    d = {}
    d['Document'] = f'{count}'
    d['Name'] = i.text
    count += 1
    corpus_data.append(d)
    if count>50:
        break
```

```python
filename = 'corpus.csv'
with open(filename, 'w', newline='', encoding="utf-8") as f:
    w = csv.DictWriter(f,['Document','Name'])
    w.writeheader()
    w.writerows(corpus_data)
```
